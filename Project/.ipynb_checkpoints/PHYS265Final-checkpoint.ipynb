```python
from astroquery.simbad import Simbad
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get stars with parallax >= 100 mas (within 10 pc)
query = """
SELECT main_id, ra, dec, plx_value
FROM basic
WHERE plx_value >= 100 
"""

# Run the query
result = Simbad.query_tap(query)
if result:
    # Make a dataframe
    df = result.to_pandas()
    
    # Rename the columns to something easy to understand
    df = df.rename(columns={
        'main_id': 'Name',
        'ra': 'RA (Degrees)',
        'dec': 'Dec (Degrees)',
        'plx_value': 'Parallax'
    })
    
    # Convert parallax to distance in parsecs (Distance = 1000 / Parallax)
    df["Distance (Parsecs)"] = 1000 / df["Parallax"]
    
    # Make sure RA and Dec are numbers
    df["RA (Degrees)"] = df["RA (Degrees)"].astype(float)
    df["Dec (Degrees)"] = df["Dec (Degrees)"].astype(float)
    
    # Make a 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(projection='3d')
    
    # Plot stars
    scatter = ax.scatter(df["RA (Degrees)"], df["Dec (Degrees)"], df["Distance (Parsecs)"], 
              c=df["Distance (Parsecs)"], cmap="viridis", alpha=0.8)
    
    # Add colorbar
    cbar = fig.colorbar(scatter)
    cbar.set_label('Distance (parsecs)', fontsize=12)
    
    # Add descriptive labels
    ax.set_xlabel('Right Ascension (degrees)', fontsize=12)
    ax.set_ylabel('Declination (degrees)', fontsize=12)
    ax.set_zlabel('Distance from Sun (parsecs)', fontsize=12)
    
    # Add title
    fig.suptitle('Stars Within 10 Parsecs of the Sun', fontsize=12)
    
    # Set background grid for better depth perception
    ax.grid(True)
    
    fig.tight_layout()
    
    # Sort stars by distance
    df = df.sort_values(by="Distance (Parsecs)")
    
    # Print the table of stars with better formatting
    print(f"Total stars found within 10pc of the Sun: {len(df)}\n")

    # Define column widths for proper alignment
    name_width = 30
    dist_width = 14  # Adjusted for better spacing
    ra_width = 14  # Adjusted width
    dec_width = 14  # Adjusted width
    
    # Print header with consistent spacing
    print(f"|{'Name':{name_width}}|{'Distance (pc)':^{dist_width}}|{'RA (deg)':^{ra_width}}|{'Dec (deg)':^{dec_width}}|")
    print("|------------------------------|--------------|--------------|--------------|")


    # Print each star with consistent formatting
    for i, row in df.iterrows():
        print(f"|{row['Name']:{name_width}}|{row['Distance (Parsecs)']:{dist_width}.3f}|{row['RA (Degrees)']:{ra_width}.1f}|{row['Dec (Degrees)']:{dec_width}.1f}|")
        print("|------------------------------|--------------|--------------|--------------|")

    plt.show()
    
else:
    print("No stars found.")
```

```python
#Calculate Hour Angle Rise and Set at College Park for Stars Within 10pc

# Function to calculate Hour Angle Rise and Set
def Hour(LAT, DEC):
    cosValue = -np.tan(LAT) * np.tan(DEC)

    if cosValue < -1:
        return np.pi, 0  # Always above horizon
    elif cosValue > 1:
        return 0, 0  # Always below horizon
    else:
        plus = np.arccos(cosValue)
        minus = -plus
        return plus, minus

# Convert degrees to radians
def deg_to_rad(deg):
    return deg * (np.pi / 180)

# Convert radians to hours
def rad_to_hour(rad):
    # 2 radians = 24 hours
    return (rad * 12 / np.pi) % 24

# Latitude of College Park, MD (39)
LAT = deg_to_rad(39)


# Convert the result to a pandas DataFrame if it's not already
if not isinstance(result, pd.DataFrame):
    df = result.to_pandas()
else:
    df = result.copy()

# Rename the columns to something easy to understand
df = df.rename(columns={
    'main_id': 'Name',
    'ra': 'RA (Degrees)',
    'dec': 'Dec (Degrees)',
    'plx_value': 'Parallax'
})

# Convert parallax to distance in parsecs (Distance = 1000 / Parallax)
df["Distance (Parsecs)"] = 1000 / df["Parallax"]

# Make sure RA and Dec are numbers
df["RA (Degrees)"] = df["RA (Degrees)"].astype(float)
df["Dec (Degrees)"] = df["Dec (Degrees)"].astype(float)

# Convert declination to radians for Hour Angle calculations
df["Dec (Radians)"] = df["Dec (Degrees)"].apply(deg_to_rad)

# Calculate Hour Angle Rise and Set for each star
rise_hours = []
set_hours = []
visibility_status = []
rise_set_radians = []

for _, row in df.iterrows():
    DEC = row["Dec (Radians)"]
    rise_rad, set_rad = Hour(LAT, DEC)
    rise_hour = rad_to_hour(rise_rad)
    set_hour = rad_to_hour(set_rad)
    
    rise_hours.append(rise_hour)
    set_hours.append(set_hour)
    rise_set_radians.append((rise_rad, set_rad))
    
    if rise_rad == np.pi and set_rad == 0:
        visibility_status.append("Circumpolar")
    elif rise_rad == 0 and set_rad == 0:
        visibility_status.append("Never visible")
    else:
        visibility_status.append("Rises and sets")

# Add the calculated values to the dataframe
df["Rise Hour Angle (hours)"] = rise_hours
df["Set Hour Angle (hours)"] = set_hours
df["Visibility at College Park"] = visibility_status

# Print detailed table of results
print(f"\nTotal stars found within 10pc of the Sun: {len(df)}\n")

# Print special section highlighting declination's impact on visibility
print("College Park Latitude: 39")
print(f"Stars with declination > 39: {sum(df['Dec (Degrees)'] > 39)} (Circumpolar)")
print(f"Stars with declination < -39: {sum(df['Dec (Degrees)'] < -39)} (Never Visible)")
print(f"Stars with -39 < declination < 39: {sum((df['Dec (Degrees)'] > -39) & (df['Dec (Degrees)'] < 39))} (Rise and Set)")

# Format table with consistent spacing
name_width = 30
dist_width = 12
dec_width = 10
rise_width = 12
set_width = 12
status_width = 15

# Print header
print(f"\n|{'Name':{name_width}}|{'Distance (pc)':^{dist_width}}|{'Dec (deg)':^{dec_width}}|" + 
      f"{'Rise HA (h)':^{rise_width}}|{'Set HA (h)':^{set_width}}|{'Status':^{status_width}}|")

print(f"|{'-'*name_width}|{'-'*dist_width}|{'-'*dec_width}|{'-'*rise_width}|{'-'*set_width}|{'-'*status_width}|")

# Create separate dataframes for each visibility status
df_circumpolar = df[df["Visibility at College Park"] == "Circumpolar"].sort_values(by="Dec (Degrees)")
df_never_visible = df[df["Visibility at College Park"] == "Never visible"].sort_values(by="Dec (Degrees)")
df_rises_sets = df[df["Visibility at College Park"] == "Rises and sets"].sort_values(by="Dec (Degrees)")

# Print data rows for Circumpolar stars
for _, row in df_circumpolar.iterrows():
    print(f"|{row['Name']:{name_width}}|{row['Distance (Parsecs)']:^{dist_width}.3f}|" + 
          f"{row['Dec (Degrees)']:^{dec_width}.2f}|{row['Rise Hour Angle (hours)']:^{rise_width}.2f}|" + 
          f"{row['Set Hour Angle (hours)']:^{set_width}.2f}|{'Circumpolar':^{status_width}}|")

# Print data rows for Never visible stars
for _, row in df_never_visible.iterrows():
    print(f"|{row['Name']:{name_width}}|{row['Distance (Parsecs)']:^{dist_width}.3f}|" + 
          f"{row['Dec (Degrees)']:^{dec_width}.2f}|{row['Rise Hour Angle (hours)']:^{rise_width}.2f}|" + 
          f"{row['Set Hour Angle (hours)']:^{set_width}.2f}|{'Never visible':^{status_width}}|")

# Print data rows for Rises and sets stars
for _, row in df_rises_sets.iterrows():
    print(f"|{row['Name']:{name_width}}|{row['Distance (Parsecs)']:^{dist_width}.3f}|" + 
          f"{row['Dec (Degrees)']:^{dec_width}.2f}|{row['Rise Hour Angle (hours)']:^{rise_width}.2f}|" + 
          f"{row['Set Hour Angle (hours)']:^{set_width}.2f}|{'Rises and sets':^{status_width}}|")
```

```python
# Get stars within RA and Dec range of 0 to 45 degrees
query = """
SELECT main_id, ra, dec, plx_value
FROM basic
WHERE ra BETWEEN 0 AND 45
AND dec BETWEEN 0 AND 45
"""

# Run the query
result = Simbad.query_tap(query)

if result:
    # Make a dataframe
    df = result.to_pandas()
    
    # Rename the columns to something easy to understand
    df = df.rename(columns={
        'main_id': 'Name',
        'ra': 'RA',
        'dec': 'Dec',
        'plx_value': 'Parallax'
    })
    
    # Convert parallax to distance in parsecs (Distance = 1000 / Parallax)
    df['Distance'] = np.nan
    mask = df['Parallax'].notna() & (df['Parallax'] > 0)
    df.loc[mask, 'Distance'] = 1000 / df.loc[mask, 'Parallax']
    
    # Make sure RA and Dec are numbers
    df["RA"] = pd.to_numeric(df["RA"])
    df["Dec"] = pd.to_numeric(df["Dec"])
    
    # Drop rows with NaN values in RA or Dec
    df = df.dropna(subset=["RA", "Dec"])
    
    # Print information about the results
    print(f"Total stars found with RA and Dec between 0-45 degrees: {len(df)}")
    
    # Separate stars with and without distance information
    dist_df = df.dropna(subset=["Distance"])
    print(f"Stars with known parallax and distance in this range: {len(dist_df)}")
    
        
else:
    print("No stars found in the specified RA and Dec range.")
```
